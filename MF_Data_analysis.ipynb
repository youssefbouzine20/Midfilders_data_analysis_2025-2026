```python
# Import Data Analysis and Visualization Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os as os
```

```python
# Load the datasets
Full_data = pd.read_csv('players_data-2025_2026.csv')
Full_data_light = pd.read_csv('players_data_light-2025_2026.csv')

# We will use the light version of the dataset for our analysis as it contains the essential features needed.
```

```python
# Filter the dataset to include only Midfielders (MF)
data_light = Full_data_light[Full_data_light['Pos'] == 'MF']
data_light

# I will use this filtered dataset for my analysis as it contains only midfielders.
```

```python
# Keep necessary columns for analysis
light_cols = [
    # ------------------Basic Player Information ------------------
    "Player", "Nation", "Pos", "Squad", "Comp", "Age", "Born",
    
    # ------------------ Playing Time & Appearances ------------------
    "MP", "Starts", "Min", "90s",
    
    # ------------------ Passing & Creativity Stats ------------------
    "PrgP", "PrgC", "KP", "Cmp%", "Ast", "xA", "PPA",
    
    # ------------------ Possession & Ball Control ------------------
    "Touches", "Carries", "PrgR", "Mis", "Dis"
]

data_light = data_light[light_cols]
data_light.head()
```

```python
# Check for missing values in the filtered dataset
missing = data_light.isnull().sum()
missing_cols = missing[missing > 0]
missing_cols
```

```python
# Fill missing values with 0 for simplicity
data_light.loc[:, ['Min','Cmp%']] = data_light[['Min','Cmp%']].fillna(0)
data_light
```

```python
# Fill missing values in Nation with 'Unknown'
data_light.loc[:,'Nation'] = data_light.fillna('Unknown')
data_light
```

```python
print(data_light.isnull().sum())
```

```python
# Extract the last country name from the Nation column
data_light['Nation'] = data_light['Nation'].apply(lambda x: x.split()[-1]).reset_index(drop=True)

data_light
```

```python
# Top passer (by KP[passes leading to goals]) for each Nation and ther squad
top_passers_per_Nation  = (
    data_light
    .sort_values(['Nation','KP'], ascending=[True, False])  # sort within each squad
    .groupby('Nation')                                      # group by squad
    .head(1)                                               # take top passer
    .reset_index(drop=True)                                # clean index
)

# Show only useful columns, sorted by KP (best to worst)
top_passers_per_Nation = top_passers_per_Nation[['Nation','Player','Comp','Squad','KP']].sort_values('KP', ascending=False).reset_index(drop=True)

top_passers_per_Nation 

```

```python
# Plot horizontal bar chart
plt.figure(figsize=(12,20))
plt.barh(top_passers_per_Nation ['Nation'], top_passers_per_Nation ['KP'], color='skyblue',)

# Add labels
plt.xlabel("Key Passes (KP)")
plt.ylabel("Squad")
plt.title("Top Passer (by KP) in Each Nation")

# Invert y-axis so the highest KP is on top
plt.gca().invert_yaxis()

# Add player names on the bars
for i, player in enumerate(top_passers_per_Nation ['Player']):
    plt.text(top_passers_per_Nation ['KP'].iloc[i] + 0.5, i, player, va='center')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# we will creat a point system to simplify clasification
data_light.loc[:,'points'] = 2*data_light['KP']+data_light['Ast']+2*data_light['Ast']+data_light['Cmp%']-data_light['Mis']-0.5*data_light['Dis']
greatest = data_light[['Age','Player','Nation','Squad','points']].sort_values(by='points',ascending=False).reset_index()
greatest=greatest.drop('index',axis=1)
```

```python
#Best 5 players 
best = greatest.head(5)

plt.figure(figsize=(10,6))
plt.scatter(best['Player'], best['points'], color=['green','red','orange','blue','purple'],)  # simple vertical bars

# Add labels and title
plt.xlabel('Player')
plt.ylabel('Points')
plt.title('Top 5 Players by Points')

plt.show()
```

```python
# Who are the best Nations ?
Nations_light = data_light.groupby('Nation')['points'].sum().sort_values(ascending=False)
Nations_light
```

```python
# Top 4 players in the best 9 Nations
fig, axes = plt.subplots(3, 3, figsize=(14,10))
axes = axes.flatten()

Best_Nations = ["ESP","FRA","GER","ENG","ITA","ARG","NED","SUI","POR"]

New_Data = greatest[greatest['points']>0]

for i,nation in enumerate (Best_Nations) :
    ax= axes[i]
    data = New_Data[New_Data['Nation']==nation]
    data = data.sort_values(by="points", ascending=False).head(4)
    ax.pie(
        data["points"],
        labels=data["Player"],
        autopct="%.2f%%",
        pctdistance=0.7,
        startangle=90
        )
    ax.set_title(nation)
    plt.suptitle("The best 4 medfilders in nevery Nation")
plt.tight_layout()
plt.show()



```

```python
# Progressive passes and there relation to the key passes leading to shoots

# --- Calculate efficiency of progressive passes for each player ---
data_light['PrgP_to_KP_ratio'] = data_light['KP'] / data_light['PrgP'] #how many progressive passes lead to key passes

# --- Replace infinite values (if PrgP=0) with 0 ---
data_light['PrgP_to_KP_ratio'] = data_light['PrgP_to_KP_ratio'].fillna(0)

plt.figure(figsize=(10,6))
plt.scatter(data_light['PrgP'], data_light['KP'], color='blue', alpha=0.6)
plt.xlabel('Progressive Passes (PrgP)')
plt.ylabel('Key Passes (KP)')
plt.title('Progressive Passes vs Key Passes for All Players')
plt.grid(True)
plt.show()



```

```python
# Top 4 players in each league
the_leagues = data_light['Comp'].unique()

fig, axes = plt.subplots(2, 3, figsize=(14, 10))
axes = axes.flatten()

for i, league in enumerate(the_leagues[:6]):  # limit to 6 leagues for 2x3 grid
    ax = axes[i]
    new_data = data_light[data_light['Comp'] == league]
    top4 = new_data.nlargest(4, "points")  # top 4 players by points

    ax.bar(top4["Player"], top4["points"], color = ["#999999", "#66c2a5", "#fc8d62", "#8da0cb"])
    ax.set_title(f"{league}")  # set league-specific title
    ax.set_xlabel("Player")
    ax.set_ylabel("Points")
    ax.tick_params(axis="x", rotation=45)


    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()



```

```python
Squad_light = data_light.groupby('Squad')['points'].sum().sort_values(ascending=False)
Squad_light
```

```python
#Best Squad Based on midfilders performance
temp_squad = Squad_light.head(5).reset_index()  # Convert Series to DataFrame
plt.pie(temp_squad['points'], labels=temp_squad['Squad'], startangle=90, autopct='%1.1f%%')
plt.title('Best Squad Based on midfilders performance')
plt.tight_layout()
plt.show()
```

```python
#Passing accuracy and its relation to the number of key passes
plt.figure(figsize=(10,6))
plt.scatter(data_light["Cmp%"], data_light["KP"], alpha=0.7, c=data_light["KP"], cmap="viridis")
for i, row in data_light.nlargest(3,"KP").iterrows():#select Top 3 players with the most key passes
    plt.text(row["Cmp%"], row["KP"], row["Player"], fontsize=8)

plt.xlabel("Pass Completion %")
plt.ylabel("Key Passes")
plt.title("Passing accuracy and its relation to the number of key passes")
plt.colorbar(label="Key Passes")
plt.tight_layout()
plt.show()
```

```python
#Expected assists (xA) compared to the actual assists achieved

plt.figure(figsize=(8,6))

# Simple scatter with regression line (optional)
sns.scatterplot(data=data_light, x="xA", y="Ast")
plt.plot([0, data_light["xA"].max()], [0, data_light["Ast"].max()], "k--")  # equality line

plt.xlabel("Expected Assists (xA)")
plt.ylabel("Actual Assists")
plt.title("Expected assists (xA) vs Actual Assists")
plt.show()


```

```python
#Passes into penalty area (PPA) and their contribution to attacking play
plt.figure(figsize=(8,6))

# Scatter plot with color = Key Passes
sns.scatterplot(
    data=data_light,
    x="PPA",
    y="Ast",
    hue="KP",
    palette="coolwarm",
    size="KP",
    sizes=(20,150),
    alpha=0.7
)

plt.xlabel("Passes into Penalty Area (PPA)")
plt.ylabel("Assists")
plt.title("Passes into the penalty area (PPA) and their contribution to attacking play")
plt.legend(title="Key Passes", bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(8,6))

# Scatter plot with regression line
sns.regplot(data=data_light, x="PrgR", y="PrgP", scatter_kws={"s":60, "alpha":0.6}, line_kws={"color":"red", "lw":2})

# Annotate top 5 players with highest PrgR
for i, row in data_light.nlargest(5, "PrgR").iterrows():
    plt.text(row["PrgR"]+0.1, row["PrgP"]+0.1, row["Player"], fontsize=8)

plt.xlabel("Progressive Receiving (PrgR)")
plt.ylabel("Progressive Passes (PrgP)")
plt.title("Progressive Receiving vs Progressive Passing – Who advances the ball more?")
plt.show()
```

```python
# Pass Completion % vs Key Passes – Who is the safe passer vs creative risk-taker? (Bubble Plot)

# Select top 10 by KP from the full data_light to get all columns needed
stats_top10 = (
    data_light
    .sort_values("KP", ascending=False)
    .head(10)
    .set_index("Player")
    [["Cmp%", "KP", "Ast", "xA", "PrgP", "PrgC", "PrgR", "Mis", "Dis"]]
)

plt.figure(figsize=(10,6))
sns.heatmap(
    stats_top10,
    annot=True, 
    fmt=".1f",
    cmap="Oranges",
    linewidths=0.5
)
plt.title("Top 10 Key Passers//Stats Heatmap")
plt.show()

```

```python
# Age and the ability of creating a key passing oportunity - Experience and creativity
plt.figure(figsize=(12,10))
sns.boxplot(data=data_light, x="Age", y="KP", palette="viridis")
plt.title("Age vs Key Passes – Does experience improve creativity?")
plt.xlabel("Age")
plt.ylabel("Key Passes")
plt.tight_layout()
plt.show()
```
